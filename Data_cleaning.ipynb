```python
def getrealizedvar(df):
    """
    This function computes the daily realized variance using the high frequency data.
    """
    realvar = {}
    Date = df['DATE'].unique()
    Symbol = df['SYMBOL'].unique()
    for symbol in Symbol:
        eachsymbol ={}
        for date in Date:
            wdf = df[(df['DATE']==date) & (df['SYMBOL'] == symbol)]
            rvec = np.asarray(wdf['RETURN'].fillna(0))
            rvar = np.inner(rvec, rvec)
            eachsymbol[date] = np.log(rvar)
        realvar[symbol] = eachsymbol
    return realvar
```

```python
total ={'AA':{}, 'C':{}, 'HAS':{}, 'INTC':{}, 'MSFT':{}, 'NKE':{}, 'PFE':{}, 'XOM':{}}
for df in dataframes:
    realvar = getrealizedvar(df)
    for key in list(total.keys()):
        total[key].update(realvar[key])
```
